```python
with open('../san_francisco-ca-2.html') as f:
    planning_code = '\n'.join(f.readlines())
```

```python
from langchain.text_splitter import HTMLHeaderTextSplitter
from bs4 import BeautifulSoup
from langchain_core.documents import Document
```

```python
soup = BeautifulSoup(planning_code)
```

```python
str(list(soup.body.children)[11].codeoptions['destid']).split('-', 1)[1]
```

```python
chunks = []
for c in list(soup.body.children):
    if c.name == 'div' and c.codeoptions:
        section_text = None
        if section := c.find_previous(class_='Section'):
            section_text = section.get_text()
        chunks.append(
            Document(
                page_content=c.get_text(),
                metadata={
                    "section": section_text,
                    "rid": str(c.codeoptions['destid']).split('-', 1)[1],
                }
            )
        )
            
```

```python

for c in list(soup.body.children)[:50]:
    if c.name == 'div' and c.codeoptions:
        c.find_previous('div')
```

```python
max([len(c.page_content) for c in chunks])
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)

final_chunks = splitter.create_documents([c.page_content for c in chunks], [c.metadata for c in chunks])

```
